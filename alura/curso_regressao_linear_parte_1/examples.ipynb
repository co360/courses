```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

data = pd.read_csv('consumo_cerveja.csv', thousands=',')

data.columns = ['data',
                'avg_temp',
                'min_temp',
                'max_temp',
                'mm',
                'weekend',
                'consumption']
```

```python
fig, ax = plt.subplots(figsize=(20, 6))
```

```python
data.head()
```

```python
ax = data['consumption'].plot(fontsize=14)
```

```python
sns.boxplot(data.consumption)
```

```python
sns.boxplot(data.consumption, orient='v', width=0.2)
```

```python
ax = sns.boxplot(data.consumption, orient='v', width=0.2)
ax.figure.set_size_inches(12, 6)
```

```python
ax = sns.pairplot(data, y_vars='consumption', x_vars=['min_temp', 'avg_temp', 'max_temp'])
ax.fig.set_size_inches(12, 6)
```

```python
ax = sns.pairplot(data)
ax.fig.set_size_inches(12, 6)
```

```python
ax = sns.pairplot(data, y_vars='consumption', x_vars=['min_temp', 'avg_temp', 'max_temp', 'weekend'], kind='reg')
ax.fig.set_size_inches(12, 6)
```

```python
ax = sns.jointplot(x='avg_temp', y='consumption', data=data)
```
