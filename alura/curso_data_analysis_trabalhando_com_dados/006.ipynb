```python
import pandas as pd
data = pd.read_excel('data_analysis/006.xlsx')
data.columns = ['date', 'views']
data.groupby(data.date.dt.year).sum().plot(ylim=(0, 300000))
```

```python
import seaborn as sns
data_per_year = data.groupby(data.date.dt.year).sum().reset_index()
sns.lmplot(x='date', y='views', data=data_per_year)
```

```python
sns.lineplot(x='date', y='views', data=data_per_year)
```
