```python
import pandas as pd
data = pd.read_excel('data_analysis/003.xlsx')
data.columns = ['day', 'views_per_day']
data = data.dropna()
data.index = range(len(data))
```

# Vari√¢ncia de views

```python
data.views_per_day.plot(figsize=(15,5))
```

```python
data.views_per_day.rolling(3).mean().dropna().plot(figsize=(15, 5))
```

```python
data.views_per_day.rolling(7).mean().dropna().plot(figsize=(15, 5))
```

```python
data.views_per_day.rolling(21).mean().dropna().plot(figsize=(15, 5))
```

# Aumento de views por dia

```python
views_increase = data.views_per_day.diff().dropna()
views_increase.index = range(len(views_increase))
day_views_increase = data.day[:-1]
data_views_increase = pd.DataFrame({'day': day_views_increase, 'views_increase': views_increase })
data_views_increase.plot('day', 'views_increase', figsize=(15, 5))
```

# Acelera√ß√£o de views por dia

```python
views_acceleration = views_increase.diff().dropna()
day_views_acceleration = day_views_increase[:-1]
data_views_acceleration = pd.DataFrame({'day': day_views_acceleration, 'views_acceleration': views_acceleration})
data_views_acceleration.plot('day', 'views_acceleration', figsize=(15, 5))
```
